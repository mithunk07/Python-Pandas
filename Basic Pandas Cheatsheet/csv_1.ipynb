  * **INDEXING AND SLICING DATA**
  
  *There are two types using which we can slice the data*
                 1.loc
                 2.iloc
                 
  * loc-  gets rows (or columns) with particular labels from the index.
  
  * iloc gets rows (or columns) at particular positions in the index (so it only takes integers).

  * **loc**

```python
data.loc[:,['Pclass','Name']].head(2)
```

```python
data.loc[1:4,['Name','Sex',"Age"]]
```

```python
data.loc[3,['Name','Sex']]
```

```python
data.loc[[2,4,5],["Name"]]
```

  * **iloc**--
  *iloc is used when we want to retrieve more number of columns so instead of typing the column name every time we can make use off the index position of the column*

```python
data.iloc[0]
```

```python
data.iloc[[12,34,55],[3,4,5]]
```

```python
data.iloc[[2,3,4],:]
```
