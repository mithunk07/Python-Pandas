## Drop Rows with Null Values

```python
nba.head(3)
```

```python
#drop any rows from the data frame that has NaN in it, ex you can see the 3rd row getting removed
nba.dropna().head()
```

```python
#and to make these changes permanent you have to use inplace = True,
nba.head()
```

```python
#remove the rows where all the values are equal to NaN
nba.tail()
```

```python
nba.dropna(how='all',inplace=True)
nba.tail()
```

```python
#To remove null values from a specific column
nba.dropna(subset=["Salary"]).head()
```

```python
nba.dropna(subset=["College",'Salary']).head()
```

## Fill in Null Values with the `.fillna()` Method

```python
nba.head()
```

```python
#if we give fillna=0 then all the NaN in the DF will be replaced by zero, but the College column is not a number 
#so replacing that with 0 is not a good option
nba.fillna(0).head()
```

```python
nba['Salary'].fillna(0,inplace=True)
nba.head()
```

```python
nba['College'].fillna("No College",inplace=True)
nba.head()
```

## The `.astype()` Method

```python
nba.dtypes
```

```python
nba["Salary"] = nba["Salary"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba.head(2)
```

```python
nba.dtypes
```

```python
#when you have small number of unique values in a column we can make use of category datatype, it helps in reducing the memory
#eg.like the month where we have 12 months,in this case the position column
nba["Position"].nunique()
```

```python
nba.info()
```

```python
#you can see the memory usage getting reduced from 35 to 32 kb
nba["Position"] = nba["Position"].astype("category")
nba.info()
```

## Rank Values with the `.rank()` Method

```python
#To rank the salary within the NBA
nba["Salary"].rank(ascending = False).astype("int").head()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head()
```

```python
nba.sort_values(by = "Salary", ascending = False).head(10)
```
