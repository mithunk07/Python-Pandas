```python
import pandas as pd
data=pd.read_csv("employees.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
#Here we can see that Gender has been imported as string which can be changed into a category and Senior Management can be changed
#into boolean,,also the date and login time is imported as string we can change it to date time format
data["Senior Management"] = data["Senior Management"].astype("bool")
data["Gender"] = data["Gender"].astype("category")
data.head(3)
```

```python
data.info()
```

```python
data['Start Date']=pd.to_datetime(data['Start Date'])
data['Last Login Time']=pd.to_datetime(data['Last Login Time'])
data.info()
```

```python
data.head()
```

## Filter A `DataFrame` Based On A Condition

```python
#remember to use double equals sign
data[data["Gender"]== "Male"].head()
```

```python
data[data["Salary"]>100000].head()
```

```python
data[data['Team'] != "Finance"].head()
```

```python
data[data["Start Date"]< "1987-01-01"].head()
```

```python
usable=data["Gender"]=="Female"
data[usable].head()
```

## Filter with More than One Condition

```python
#AND condition-both the condition have to be true inorder to exrtract a row.
data[(data["Gender"]=="Female") & (data['Senior Management']==True)].head()
```

```python
use1=data["Gender"]=="Female"
use2=data['Senior Management']==True
data[use1 & use2].head()
```

```python
#OR condition-both the condition have to be true inorder to exrtract a row.
use1=data["Gender"]=="Female"
use2=data['Team']=="Finance"
data[use1 | use2].head()
```

```python
use1=data["Gender"]=="Female"
use2=data['Team']=="Finance"
use3=data['First Name']=='Ruby'
data[use3 & use1 | use2].head()
```
