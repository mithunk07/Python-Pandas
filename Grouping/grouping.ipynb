## ***GROUPING***

```python
import numpy as np
import pandas as pd

data=pd.read_csv('train.csv')

data.head()
```

```python
s1=data.groupby("Sex")
s1.groups
```

```python
s1.agg(["count"])
```

```python
s1["Age","Fare"].agg(["min","max"])
```

```python
s1["Sex"].agg(["count"])
```

```python
s1=data.groupby(["Sex","Pclass"])
s1.groups
s1["Sex"].agg(["count"])
```

```python
s1=data.groupby(["Sex","Pclass","Embarked"])
s1.groups
s1["Sex"].agg(["count"])
```
