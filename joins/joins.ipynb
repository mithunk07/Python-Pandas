## ***Joins***

```python
import numpy as np
import pandas as pd


emp=pd.read_csv(r'C:\Users\MITHUNKUMAR S\Desktop\Jupyter\joins\emp.csv')
dept=pd.read_csv(r'C:\Users\MITHUNKUMAR S\Desktop\Jupyter\joins\dept.csv')
emp
```

```python
dept
```

  * ***INNER JOIN(Default)***--
  **Returns record when there is  matching values in both the dataframes**

```python
inner=pd.merge(emp, dept, on='Deptno', how='inner')
inner
```

  * ***LEFT JOIN***--**Return all records from the left dataframe, and the matched records from the right dataframe**

```python
left=pd.merge(emp, dept, on='Deptno',how='left')
left
```

* ***RIGHT JOIN***--**Return all records from the right dataframe, and the matched records from the left dataframe**

```python
right=pd.merge(emp, dept, on='Deptno',how='right')
right
```

  * ***OUTER JOIN***--**Return all records when there is a match in either left or right dataframe**

```python
outer=pd.merge(emp,dept,on="Deptno",how="outer")
outer
```
